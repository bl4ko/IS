# Data Preprocessing Tools

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

## Taking care of missing data

```python
from sklearn.impute import SimpleImputer
# 1. Create an imputer object
imputer = SimpleImputer(missing_values=np.nan, strategy='mean') # Stretegy can be mean, median, most_frequent, constant
imputer.fit(X[:, 1:3]) # Calculate the mean of the columns
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

```python
print(X)
```

## Encoding categorical data

### Encoding the Independent Variable

Encode using binary vectors (one-hot encoding)

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

### Encoding the Dependent Variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

## Feature Scaling
Required for certain models.

Two techniques:
- `Standardization` $X' = \frac{X - \mu}{\sigma}$
    - All values are between -3 and +3.
    - in all cases
- `Normalization`:  $X' = \frac{X - X_{min}}{X_{max} - X_{min}}$
    - All values are between 0 and 1.
    - recommended when most of the features follow a normal distribution.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# We don't have to transform categorical data (0, 1, 2) because they are already in the same scale
# We only need to transform the numerical data 
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])
X_test[:, 3:] = sc.transform(X_test[:, 3:])
```
